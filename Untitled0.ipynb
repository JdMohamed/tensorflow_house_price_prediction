```python
import tensorflow as tf
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

**Data Preparation**

```python
data = pd.read_csv("sample_data/california_housing_train.csv")
data.shape
data.head()
```

```python
data.columns
```

```python
sns.pairplot(data[['longitude', 'latitude', 'housing_median_age', 'total_rooms',
       'total_bedrooms', 'population', 'households', 'median_income',
       'median_house_value']],diag_kind='kde')
```

```python
tensor_data =tf.constant(data)
print(tensor_data)
```

```python
tensor_data = tf.random.shuffle(tensor_data)
print(tensor_data)
```

```python
X = tensor_data[:,:-1]
y = tensor_data[:,-1]
y = tf.expand_dims(y,axis=-1)
print(X[:5])
```

```python
DATASET_SIZE = len(X)
TRAIN_SPLIT = int(0.7 * DATASET_SIZE)
VAL_SPLIT = int(0.15 * DATASET_SIZE)
TEST_SPLIT = int(0.15 * DATASET_SIZE)
X_train = X[:TRAIN_SPLIT]
y_train = y[:TRAIN_SPLIT]
X_val = X[TRAIN_SPLIT:TRAIN_SPLIT+VAL_SPLIT]
y_val = y[TRAIN_SPLIT:TRAIN_SPLIT+VAL_SPLIT]
X_test = X[TRAIN_SPLIT+VAL_SPLIT:]
y_test = y[TRAIN_SPLIT+VAL_SPLIT:]
```

```python
train_dataset = tf.data.Dataset.from_tensor_slices((X_train,y_train))
train_dataset = train_dataset.shuffle(buffer_size=32,reshuffle_each_iteration=True).batch(512)
val_dataset = tf.data.Dataset.from_tensor_slices((X_val,y_val))
val_dataset = val_dataset.shuffle(buffer_size=32,reshuffle_each_iteration=True).batch(512)
test_dataset = tf.data.Dataset.from_tensor_slices((X_test,y_test))
test_dataset = test_dataset.shuffle(buffer_size=32,reshuffle_each_iteration=True).batch(512)
```

```python
from tensorflow.keras.layers import Normalization,Dense,InputLayer
from tensorflow.keras.losses import MeanSquaredError,MeanAbsoluteError
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.metrics import RootMeanSquaredError
```

```python

normalizer = Normalization()
normalizer.adapt(X_train)
print(normalizer(X)[:5])
```

```python
model = tf.keras.Sequential([
    InputLayer((8,)),
    normalizer,
    Dense(128,activation='relu'),
    Dense(128,activation='relu'),
    Dense(64,activation='relu'),
    Dense(1)
])
model.summary()
```

```python
model.compile(optimizer=Adam(learning_rate=0.1),loss=MeanAbsoluteError(),metrics=[RootMeanSquaredError()])
```

```python
history = model.fit(train_dataset,validation_data=val_dataset,epochs=100,verbose=0)
```

```python

plt.figure(figsize=(12,4))
plt.subplot(1,2,1)
plt.plot(history.history['loss'] , label='loss')
plt.plot(history.history['val_loss'] , label='val loss')
plt.ylabel('loss')
plt.xlabel('epochs')
plt.legend(['train','val'])
plt.subplot(1,2,2)
plt.plot(history.history['root_mean_squared_error'] , label='root mean squared error')
plt.plot(history.history['val_root_mean_squared_error'] , label='val root mean squared error')
plt.ylabel('root mean squared error')
plt.xlabel('epochs')
plt.legend(['root mean squared error','val men squared error'])
plt.show()
```

```python
model.evaluate(X_test,y_test)
```

```python
y_pred = model.predict(X_test)
print(list(y_pred[:,0])[0:10])
```
